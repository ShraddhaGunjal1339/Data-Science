```python
df.dropna(subset=["Age"],axis=0,inplace=True)
```

```python
#replace missing value
df['Age'].fillna(df['Age'].mean())
```

```python
df['Age'].replace(to_replace=np.nan, value=df['Age'].mean())
```

```python
df.replace(to_replace = np.nan, value =-99999)
```

```python
df.replace(to_replace = "Petrol", value ="Gas")
```

```python
#data formatting
df.dtypes
```

```python
df["FuelType"]=df["FuelType"].astype('string')
df.dtypes
```

```python
#single feature scaling
df["Age"]=df["Age"]/df["Age"].max()
```

```python
#min max
df['Age'] = (df['Age'] - df['Age'].min())/(df['Age'].max() - df['Age'].min())
```

```python
#Z-Score
df['Age'] = (df['Age']- df['Age'].mean())/df['Age'].std()
```

```python
#Binning
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('Toyota.csv')
min_value=min(df["Price"])
max_value=max(df["Price"])
bins=np.linspace(min_value,max_value,4)
print(bins.round(2))
labels=['Low','Medium','High'] 
df["Price_Binned"]=pd.cut(df["Price"],bins=bins,labels=labels,include_lowest=True) 
plt.hist(df["Price_Binned"], bins=3)
plt.show()
```
